# QUESTION 5

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([5, 7, 9, 7, 13])
plt.subplot(1, 3, 1)
plt.plot(x1, y1)
plt.title('Hours Sleeping')
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([1, 2, 4, 7, 11])
plt.subplot(1, 3, 2)
plt.plot(x2, y2)
plt.title('Hours Studying')
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([1, 1.5, 2, 2.5, 3])
plt.subplot(1, 3, 3)
plt.plot(x3, y3)
plt.title('GPA During Pandemic')

plt.suptitle('Covid-19 on Student')
plt.show()
```
